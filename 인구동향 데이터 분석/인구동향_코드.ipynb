```python
import csv
f = open('지역별_인구동향.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)                   # 헤더 건너띄기
home = []                    # 입력 받은 지역의 데이터를 저장할 리스트 생성
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ') # 자신이 알고싶은 지역을 검색
for row in data :
    if name in row[0] :     # 입력 받은 지역의 이름이 포함된 행 찾기,첫번째 컬럼값 중에 '동네이름'이 포함되었으면
        print(row[0])       # 입력한 지역와 출력된 지역이 맞는지 눈으로 확인
        print(row)          # 찾은 동네 행의 데이터를 그대로 출력
        for i in row[3:] : # 3번 인덱스 값부터 슬라이싱하여 끝까지 출력
            home.append(int(i)) # 입력 받은 지역의 데이터를 정수로 변환하여 home 리스트에 저장
print('0세~100세까지의 각각 인구수:' ,home)                     # home에 저장된 데이터 출력
```

```python
# 배열을 처리하기에는 Numpy 라이브러리가 효율적, numpy 버전으로 변환
import numpy as np
import csv
f =open('지역별_인구동향.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype = int) # 3번컬럼 이후의 모든 값을 정수로 변환 후 home에 저장

import matplotlib.pyplot as plt  # matplotlib의 하위 패키지 pyplot을 plt으로 별칭하여 불러온다.
plt.plot(home) # home에 있는 리스트를 그래프 그린다.
plt.show() #그래프를 그린 plt를 출력한다.
```

# 지역의 평균 연령을 구하자

```python
import csv
f =open('지역별_인구동향.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype = int) # 3번컬럼 이후의 모든 값을 정수로 변환 후 home에 저장
average_age = 0
sum_people = 0
for i in range(100):
    average_age += i*home[i]
    sum_people += home[i]

print('평균연령 : ',average_age/sum_people)
```

# 인구구조를 인구밀도로 표현한다
## 지역마다 인구수가 다른 것을 인구 밀도로 정규화를 시켜 비교한다.

```python
# 나이별 인구 비율 구하기
import numpy as np
import csv
f =open('지역별_인구동향.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
data = list(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype=int) / int(row[2])  # 인구수/전체인구수 로 하여 전체적인 정규화를 진행한다.

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)    
plt.rc('font', family ='AppleGothic')
plt.title(name +' 지역의 인구 구조')

plt.plot(home) 
plt.show()
```

#### A지역과 차이가 가장 적은 B 지역을 찾는다. 왜? 인구 구조가 상이한 지역이 유사한 지역으로 나올까?

```python
import numpy as np
import csv
f =open('지역별_인구동향.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
data = list(data)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
mn = 1           # 최솟값을 저장할 변수 생성 및 초기화(최대 1)
result_name = '' # 최솟값을 갖는 지역의 이름을 저장할 변수 생성 및 초기화
result = 0       # 최솟값을 갖는 지역의 연령대별 인구 비율을 저장할 배열 생성 및 초기화
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype = int) / int(row[2]) 
for row in data : 
    away = np.array(row[3:], dtype = int) / int(row[2]) 
    s = np.sum(home - away) # (원하는지역-away지역)을 합친다.
    if s < mn : 
        mn = s 
        result_name = row[0] # 가장 차이가 적은 인구구조를 가진 지역명을 저장
        result = away # 가장 차이가 적은 인구구조를 저장
            
import matplotlib.pyplot as plt
plt.plot(home)
plt.plot(result)
plt.savefig('result.png')
print(result_name)
plt.show()
```

#### A 지역과 인구 구간별 비율 차이가 가장 적은 B 지역을 찾기 위해 차이를 제곱한다.

```python
import numpy as np
import csv
f =open('지역별_인구동향.csv','r',encoding='cp949')
data = csv.reader(f) # data는 1개의 객체 덩어리이다. for문으로 읽으면 커서가 끝까지 가서, 다시 for문으로 읽을 수 없다
next(data)
data = list(data)    # 그래서 data를 리스트로 만든다.
# 궁금한 지역의 이름을 입력받는다.
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
mn =1
result_name =''
result =0
# 궁금한 지역의 인구 구조를 저장한다.
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype =int) /int(row[2])
# 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역을 찾는다.
for row in data :
    away = np.array(row[3:], dtype =int) /int(row[2])
    s = np.sum((home - away) ** 2)          # 제곱하여 실제 간격이 최소인 것을 찾는다.
    if s < mn and name not in row[0] :
        mn = s
        result_name = row[0]
        result = away
            
# 궁금한 지역의 인구 구조와 가장 비슷한 곳의 인구 구조를 시각화한다.
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)            
plt.rc('font', family ='AppleGothic')
plt.title(name +' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label = name)
plt.plot(result, label = result_name)
plt.legend()
plt.savefig('result_result.png')
plt.show()
```
